HEATING is the most common complaint_type in the dataset, let's take a look deeper

```python
#How often each complaints appears in the data? Calculating the frequency
complaintTypes_frequency = df['complaint_type'].value_counts()
complaintTypes_frequency
```

From the most to the least frequent complaint types. Heating is the mots common, followed by General Construction, plumbing, paint-plaster, ... Therefore the Department of housing should focus first on the complainsthat are more common.

However we can see that there are two complains that can be related one to the other. These are the heating and heat/hot water. Probably a heating/heat problem can be coming from not having hot water. Therefore is we sum the number of heating and heat/hot water complaint, this would be around 503806, a 33% of the complains.

```python
#Summation of the frequencies for heating and heat/hot water
complaintTypes_frequency['HEATING']+complaintTypes_frequency['HEAT/HOT WATER']
```
